```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

alumnos =  pd.read_csv("Alumnos.csv", encoding='latin-1')
print (alumnos)
#print(alumnos['Nombre'])
alumnos['Nombre Completo'] = alumnos['Nombre'] + ' ' + alumnos['Apellido']
print(alumnos)
np.random.seed(42)
```

```python
datos={
    "Nombre Completo": alumnos['Nombre Completo'],
    "Edad":np.random.randint(18,25,size=27),
    "Calificación":np.random.normal(7.5,1.0,size=27).round(1),
    "Asistencia": np.random.uniform(70,100,size=27).round(1)
}
df = pd.DataFrame(datos)
print(df)
```

```python
plt.hist(df["Edad"], bins=6, edgecolor="black", color='pink')
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribución de Edades")
plt.grid(axis='y',alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(df["Nombre Completo"], df["Asistencia"], color='orchid',edgecolor='black')
plt.xlabel("Alumno")
plt.ylabel("Asistencia (%)")
plt.title("Porcentaje de asistencia por alumno")
plt.xticks(rotation=90)
plt.ylim(50,100)
plt.grid(axis='y', linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df["Edad"], df["Asistencia"], color='violet')
plt.xlabel("Edad")
plt.ylabel("Asistencia (%)")
plt.title("Relación entre Edad y Porcentaje de Asistencia")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(df["Nombre Completo"], df["Calificación"], color='navy',edgecolor='black')
plt.xlabel("Alumno")
plt.ylabel("Calificación")
plt.title("Calificaciones de los alumnos")
plt.xticks(rotation=90)
plt.ylim(5,10)
plt.grid(axis='y', linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()
```
