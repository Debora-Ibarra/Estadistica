### Ejercicio 1
En una muestra de 105 comercios seleccionados al azar de una zona, se
observa que 27 de ellos han tenido pérdidas en este mes. Un analista
económico de la zona establece que la proporción de comercios en la zona
con pérdidas es igual o superior al 35%. Realice una prueba de hipótesis con
un nivel de significancia del 5% para dicha afirmación e interprete sus
resultados.

$H_0:\mu_0\geq35\%$

$H_1:\mu_0<35\%$

```python
import scipy.stats as ss
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Datos
num_datos1 = 105
mu_01 = 0.35
mu1 = 27 / 105
alpha = 0.05
```

```python
# Estadistico z
estadistico_z1 = (mu1 - mu_01) / np.sqrt((mu_01 * (1 - mu_01)) / num_datos1)
```

```python
# P-value cola izquierda
p_value1 = ss.norm.cdf(estadistico_z1)
```

```python
# Imprimir
print(f't = {estadistico_z1}\np_value = {p_value1}')
if p_value1 < alpha:
    print("Se rechaza la hipótesis nula en favor de la hipótesis alternativa")
else:
    print("No se tiene evidencia suficiente para rechazar a la hipótesis nula")
```

```python
#Crear función
x1 = np.linspace(-4, 4, 1000)
y1 = ss.norm.pdf(x1)
z_critico = ss.norm.ppf(1 - alpha)
```

```python
# Graficar
plt.plot(x1, y1, label=f'Distribución normal', color='blue')
plt.fill_between(x1, 0, y1, where=(x1 <= -z_critico), color='red', alpha=0.3)
plt.axvline(estadistico_z1, color='red', linestyle='--', alpha=0.5, label=f'Estadístico z = {estadistico_z1}')
plt.axvline(-z_critico, color='black', linestyle=':')

plt.title('Prueba normal (cola izquierda)')
plt.xlabel('Valor z')
plt.ylabel('Densidad')
plt.legend()
plt.grid()
plt.show()
```

#### Ejercicio 2
Una encuesta afirma que el 50% de los votantes apoyan a cierto candidato.
Sin embargo, un investigador desea verificar si la proporción real de apoyo
difiere significativamente de 50%. Se toma una muestra aleatoria de 200 votantes y se observa que 112 de ellos apoyan al candidato.¿Hay evidencia para afirmar que la proporción real de apoyo es distinta de 50%? Usa un nivel de significancia del 5%

$H_0:\mu_0=50\%$

$H_1:\mu_0\neq50\%$

```python
# Datos
num_datos2 = 200
mu_02 = 0.5
mu2 = 112 / 200
```

```python
# Estadistico z
estadistico_z2 = (mu2 - mu_02) / np.sqrt((mu_02 * (1 - mu_02)) / num_datos2)
```

```python
# P-value dos colas
p_value2 =  2 * (ss.norm.cdf(estadistico_z1))
```

```python
# Imprimir
print(f't = {estadistico_z2}\np_value = {p_value2}')
if p_value2 < alpha:
    print("Se rechaza la hipótesis nula en favor de la hipótesis alternativa")
else:
    print("No se tiene evidencia suficiente para rechazar a la hipótesis nula")
```

```python
#Crear función
x2 = np.linspace(-4, 4, 1000)
y2 = ss.norm.pdf(x2)
z_critico = ss.norm.ppf(1 - alpha)
```

```python
# Graficar
plt.plot(x2, y2, label=f'Distribución normal', color='blue')
plt.fill_between(x2, 0, y2, where=(x2 >= z_critico), color='red', alpha=0.3, label=f'Región crítica (α/2={alpha/2})')
plt.fill_between(x2, 0, y2, where=(x2 <= -z_critico), color='red', alpha=0.3)
plt.axvline(estadistico_z2, color='red', linestyle='--', alpha=0.5, label=f'Estadístico z = {estadistico_z2}')
plt.axvline(z_critico, color='black', linestyle=':', label=f'Valores críticos = ±{z_critico:.3f}')
plt.axvline(-z_critico, color='black', linestyle=':')

plt.title('Prueba normal (dos colas)')
plt.xlabel('Valor z')
plt.ylabel('Densidad')
plt.legend()
plt.grid()
plt.show()
```

### Ejercicio 3
Un ingeniero de calidad afirma que la varianza en el peso de un producto no
debe superar 0.04 kg2 para garantizar la uniformidad del empaque. Se toma
una muestra aleatoria de 20 productos y se obtiene una varianza muestral de
0.065 kg2. ¿Existe evidencia suficiente para concluir que la varianza poblacional es mayor a 0.04? Usa un nivel de significancia de 5%.

$H_0:\mu_0\leq0.04$

$H_1:\mu_0>0.04$

```python
# Datos
var_pob = 0.04
num_datos3 = 20
var_mues = 0.065
```

```python
# Estadistico ji
ji = (num_datos3 - 1) * var_mues / var_pob
```

```python
# P-value cola derecha
p_value3 =  1 - ss.chi2.cdf(ji, num_datos3 - 1)
```

```python
# Imprimir
print(f'ji = {ji}\np_value = {p_value3}')
if p_value3 < alpha:
    print("Se rechaza la hipótesis nula en favor de la hipótesis alternativa")
else:
    print("No se tiene evidencia suficiente para rechazar a la hipótesis nula")
```

```python
#Crear función
x3 = np.linspace(0,40, 1000)
y3 = ss.chi2.pdf(x3, num_datos3 - 1)
ji_critico = ss.chi2.ppf(1 - alpha, num_datos3 - 1)
```

```python
#Graficar
plt.plot(x3, y3, label=f'Distribución χ2 (gl = {num_datos3 - 1})', color='blue')
plt.fill_between(x3, 0, y3, where=(x3 >= ji_critico), color='red', alpha=0.3, label='Región crítica (α=0.05)')
plt.axvline(ji_critico, color='black', linestyle=':', label=f'Valor crítico = {ji_critico:.2f}')
plt.axvline(ji, color='red', linestyle='--', label=f'Estadístico ji = {ji:.2f}')


plt.title('Prueba χ2 (cola derecha)')
plt.xlabel('Valor χ')
plt.ylabel('Densidad')
plt.legend()
plt.grid()
plt.show()
```
