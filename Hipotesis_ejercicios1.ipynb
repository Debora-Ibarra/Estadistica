#### Ejercicio 1
Un analista afirma que el rendimiento promedio mensual de una acción es superior al 1%. Para probarlo toma una muestra de 12 meses la cual tiene una media de 1.4% y la desviación estándar muestral es de 0.6%. Considerando un nivel de significancia del 5%, realice una prueba de hipótesis y verifique está afirmación.

$H_0:\mu_0>0.01$

$H_1:\mu_0\leq0.01$

```python
import scipy.stats as ss
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Datos
mu = 0.014
mu_0 = 0.01
desvest = 0.006
num_datos = 12
alpha = 0.05
```

```python
# Estadistico T
estadistico_t = (mu - mu_0) / (desvest / np.sqrt(num_datos))
```

```python
# P-value cola derecha
p_value = 1 - ss.t.cdf(estadistico_t, num_datos - 1)
```

```python
# Imprimir
print(f't = {estadistico_t}\np_value = {p_value}')
if p_value < alpha:
    print("Se rechaza la hipótesis nula en favor de la hipótesis alternativa")
else:
    print("No se tiene evidencia suficiente para rechazar a la hipótesis nula")
```

```python
#Crear función
x = np.linspace(-4, 4, 1000)
y = ss.t.pdf(x, num_datos - 1)
t_critico = ss.t.ppf(1 - alpha, num_datos - 1)
```

```python
#Graficar
plt.plot(x, y, label=f'Distribución t (gl = {num_datos - 1})', color='blue')
plt.fill_between(x, 0, y, where=(x >= t_critico), color='red', alpha=0.3, label='Región crítica (α=0.05)')
plt.axvline(t_critico, color='black', linestyle=':', label=f'Valor crítico = {t_critico:.2f}')
plt.axvline(estadistico_t, color='red', linestyle='--', label=f'Estadístico t = {estadistico_t:.2f}')


plt.title('Prueba t (cola derecha)')
plt.xlabel('Valor t')
plt.ylabel('Densidad')
plt.legend()
plt.grid()
plt.show()
```

#### Ejercicio 2
Un analista quiere verificar si un fondo tiene un rendimiento diferente del promedio de mercado (6%). Se realiza un muestreo de 30 datos los cuales tienen una media muestral de 5.2% y desviación estándar muestral de 1.8% con un nivel de significancia del 5%. Realice una prueba de hipótesis y verifique si se cumple lo que el analista plantea.

$H_0:\mu_0=0.06$

$H_1:\mu_0\neq0.06$

```python
# Datos
num_datos2 = 30
mu_02 = 0.06
mu2 = 0.052
desvest2 = 0.018
```

```python
# Estadistico T
estadistico_t2 = (mu2 - mu_02) / (desvest2 / np.sqrt(num_datos2))
```

```python
# P-value dos colas
p_value2 =  2 * (1 - ss.t.cdf(estadistico_t, num_datos - 1))
```

```python
# Imprimir
print(f't = {estadistico_t2}\np_value = {p_value2}')
if p_value2 < alpha:
    print("Se rechaza la hipótesis nula en favor de la hipótesis alternativa")
else:
    print("No se tiene evidencia suficiente para rechazar a la hipótesis nula")
```

```python
#Crear función
x2 = np.linspace(-4, 4, 1000)
y2 = ss.t.pdf(x2, num_datos2 - 1)
t_critico2 = ss.t.ppf(1 - alpha, num_datos2 - 1)
```

```python
# Graficar
plt.plot(x2, y2, label=f'Distribución t (gl={num_datos2})', color='blue')
plt.fill_between(x, 0, y, where=(x >= t_critico2), color='red', alpha=0.3, label=f'Región crítica (α/2={alpha/2})')
plt.fill_between(x, 0, y, where=(x <= -t_critico2), color='red', alpha=0.3)
plt.axvline(estadistico_t2, color='red', linestyle='--', alpha=0.5, label=f'Estadístico t = {estadistico_t2}')
plt.axvline(t_critico2, color='black', linestyle=':', label=f'Valores críticos = ±{t_critico2:.3f}')
plt.axvline(-t_critico2, color='black', linestyle=':')

plt.title('Prueba t (dos colas)')
plt.xlabel('Valor t')
plt.ylabel('Densidad')
plt.legend()
plt.grid()
plt.show()
```
